```python
#who will pay the bill?
import random
a=["latha","charu","kevin","sanju","varsha","deborah"]
b=random.choice(a)
print(b)
# either by both you can use
z=random.randint(0,4)
if(z==0):
    print(a[0])
elif(z==1):
    print(a[1])
elif(z==2):
    print(a[2])
elif(z==3):
    print(a[3])
elif(z==4):
    print(a[4])
# else you could also try
friends=random.randint(0,4)
print(a[friends])
```
