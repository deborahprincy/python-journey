```python
#rock,paper,sicissors game
import random
rock=(''' ROCK
_______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
''')
paper=('''PAPER
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
''')
scissors=('''SCISSORS
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
''')
choice=[rock,paper,scissors]
user=int(input('''What you choose to select? \n TYPE\n 0 for rock \n 1 for paper \n 2 for scissors'''))
if(user<0 or user>2):
    print("invalid input")
else:
    print("your choice is: ",choice[user])
    computer=random.randint(0,2)
    print("computer choice is: ",choice[computer])

    if(user == computer):
        print("it's a draw")
    elif((user==1 and computer==0) or (user==0 and computer==2) or (user==2 and computer==1) ):
         print("it's a win")
    else:
        print("you lose")
    
    
               
               

```
