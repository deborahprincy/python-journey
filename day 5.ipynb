```python
# assignment work( get to find the max of the list without using max()
num=[199999999,2,33000,123,5,6000000,7,8,4556]
max_score=0
for i in num:
    if(i>max_score):
        max_score=i
print(max_score)

```

```python
# assignment work( get to find the reverse of the list without using reverse()
num=[199999999,2,33000,123,5,6000000,7,8,4556]
reverse_list=[]
for i in range((len(num)-1),-1,-1):
    reverse_list.append(num[i])
print(reverse_list)
```
